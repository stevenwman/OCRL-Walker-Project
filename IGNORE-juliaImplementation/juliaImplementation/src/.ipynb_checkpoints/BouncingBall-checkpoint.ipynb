# Iterative LQR for Hybrid State Estimation
Implementation of iterative LQR (ilQR) for state estimation, using the Gauss-Newton approximation of differential dynamic programming (DDP). Used to estimate trajectories for a bouncing ball model.

## The Model

$$ 
x = \begin{bmatrix} p_x \\ p_y \\ v_x \\ v_y \end{bmatrix}, \quad
\dot{x} = \begin{bmatrix} v_x \\ v_y \\ 
0 \\
- g \\
\end{bmatrix}
$$

```julia
const nx = 4 
const ny = 2

# const Q = Diagonal(([ones(3) ; fill(0.1, 3)]))
# const R = Diagonal(fill(1e-2, nu))
# const Qf = Diagonal(fill(1e2, nx));
const C = [I zeros(2,2)]
# const C = Matrix(I,4,4)
# const V = Diagonal([.01,.01,.01,.01])
const V = Diagonal([.01,.01])
const W = Diagonal([.1,.1,1,1])

const dt = 0.025 
const N = 301

# lqrparam = LQRParameters(nx,nu,Q,R,Qf,dt,N)
estparam = ESTParameters(nx,ny,C,V,W,dt,N);
```

## Simple Trajectory
Functions to simulate the trajectory and sensor readings from a noisy system

```julia
function ballDynamics(dx,x,p,t)
    dx[1] = x[3]+p[2][1]
    dx[2] = x[4]+p[2][2]
    dx[3] = 0+p[2][3]
    dx[4] = p[1]+p[2][4]
end

function dynamics_jacobians(x,w,dt)
    dx = zeros(Float64,4,4)
    dx[1,3] = dt
    dx[2,4] = dt
    A = Matrix(I,4,4)+dx
    B = Matrix(I,4,4)*dt
    return A,B
end

function groundEvent(x,t,integrator)
    x[2]
end

function resetMap!(integrator)
    integrator.u[4] = -0.8*integrator.u[4]
end

function dynamicStep(X,dt,p)
    bouncecb = ContinuousCallback(groundEvent,resetMap!)
    prob = ODEProblem(ballDynamics,X,(0,dt),p)
    sol = solve(prob,Tsit5(),callback=bouncecb)
    impactstate = deepcopy(X)
    postimpactstate = deepcopy(X)
    impact = 0
    for i = 1:size(sol,2)
        if abs(sol[i][2]) < 1e-10 && sol[i][4] < 0
            impactstate = sol[i]
            postimpactstate = sol[i+1]
            impact = 1
        end
    end
    return sol[end],impact,impactstate, postimpactstate
end

function getTrajectory(x0,dt,g,N,C,W,V)
    X = [x0 for i = 1:N]
    Y = [zeros(size(C)[1]) for i = 1:N]
    Y[1] = C*x0 + sqrt(V)*randn(size(Y[1]))
    impactstates = [x0 for i = 1:N-1]
    postimpactstates = [x0 for i = 1:N-1]
    impacts = [false for i = 1:N-1]
    for i = 1:N-1
        noise = sqrt(W)*randn(size(X[i]))
        p = [g,noise]
        X[i+1],impacts[i],impactstates[i],postimpactstates[i] = dynamicStep(X[i],dt,p)
        Y[i+1] = C*X[i+1] + sqrt(V)*randn(size(Y[1]))
    end
    return X,Y,impacts,impactstates, postimpactstates
end
    
```

## Saltation Matrix Related Code
Need f1,f2,r,DxR,dxg, dtg

```julia
function r1(x,p,t)
    x2 = x
    x2[4] = -0.8*x[4]
    return x2
end

function get_DxR(x,p,t)
    DxR = Diagonal([1,1,1,-0.8])
    return DxR
end

function get_Dxg(x,p,t)
    Dxg = [0;1;0;0]
    return Dxg
end

function saltation(x,p,t)
    f1 = deepcopy(x)
    ballDynamics(f1,x,p,t)
    xp = r1(x,p,t)
    f2 = deepcopy(xp)
    ballDynamics(f2,xp,p,t)
    DxR = get_DxR(x,p,t)
    Dxg = get_Dxg(x,p,t)
    salt = DxR + ((f2-DxR*f1)*Dxg')/(Dxg'*f1)
    return salt
end
```

```julia
x = [1,0,.5,-10]
noise = [0,0,0,0]
g = -9.8
p = [g,noise]
t = 0
S = saltation(x,p,t)
```

```julia
# reference trajectory (we just want to track the goal state and hover control)
x0 = [0,10,.5,0]
g = -9.8


Xnoisy, Ynoisy, impacts, impactstates, postimpactstates = getTrajectory(x0,dt,g,N,C,W,V)
X_nonoise, Yno, imp, imps, postimps = getTrajectory(x0,dt,g,N,C,.0000001*W,V)
Xnoisy = SVector{nx}.(Xnoisy)
Ynoisy = SVector{ny}.(Ynoisy)
Xno = SVector{nx}.(X_nonoise)
# plotlyjs()
traj2(Xnoisy, label="Simulation with noise")
traj2!(Ynoisy,label="Sensor readings")
traj2!(Xno)
```

## Estimate the Trajectory

```julia
Ww = 0*Xnoisy[2:end]
Xest,West,K,P,iter = iLQR_HMHE(estparam,x0,Ww,Ynoisy,max_iters=20);
```

```julia
Xest = SVector{nx}.(Xest)
traj2(Xnoisy, label="Trajectory",dpi=600)
traj2!(Xest, label="Estimation")
traj2!(Ynoisy,label="Measurements")
savefig("BouncingBall.png")
```

```julia
Xn = [Xnoisy[i] for i=1:301]
Xe = [Xest[i] for i=1:301]
times = range(0, step=dt, length=N)
plot(times,Xn,layout=4,dpi=600,xlabel="time [s]",label=["True x" "True y" "True xÌ‡" "True yÌ‡"])
plot!(times,Xe,layout=4,label=["Est x" "Est y" "Est xÌ‡" "Est yÌ‡"])

savefig("BouncingBall_Elementwise.png")
```

```julia
# Running lots of trials and generating average plots:
# reference trajectory (we just want to track the goal state and hover control)
x0 = [0,10,.5,0]
g = -9.8

n_trials = 100

avg_error = 0*Xest

for x in 1:n_trials
    Xnoisy, Ynoisy, impacts, impactstates, postimpactstates = getTrajectory(x0,dt,g,N,C,W,V)
    Ww = 0*Xnoisy[2:end]
    Xest,West,K,P,iter = iLQR_HMHE(estparam,x0,Ww,Ynoisy,max_iters=20);
    scaled_error = (Xest.-Xnoisy)./n_trials
    for j = 1:size(scaled_error,1)
        scaled_error[j] .= abs.(scaled_error[j])
    end
    
    avg_error = avg_error .+ scaled_error
end
avg_error = SVector{nx}.(avg_error)
avg_error = [avg_error[i] for i=1:301]
times = range(0, step=dt, length=N)
plot(times,avg_error,layout=4,dpi=600,xlabel="time [s]",label=["x error" "y error" "xÌ‡ error" "yÌ‡ error"])

savefig("BouncingBall_AvgError.png")

# Xnoisy = SVector{nx}.(Xnoisy)
# Ynoisy = SVector{ny}.(Ynoisy)
# Xno = SVector{nx}.(X_nonoise)
# plotlyjs()
```
