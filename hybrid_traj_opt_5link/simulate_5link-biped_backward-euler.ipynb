```julia
include("DynamicSystems/biped5link.jl") # Defines hybrid system functions

model = ( m1 = 1,  m2 = 1,  m3 = 1,  m4 = 1,  m5 = 1, m6 = 1,
		      l12 = 1, l23 = 1, l34 = 1, l45 = 1, l36 = 1, g = 9.81)


foot1_pos = [0, 0]

θ1 = 280 * (π/180)
θ2 = 300 * (π/180)
θ3 = 90 * (π/180)
θ4 = 250 * (π/180)
θ5 = 220 * (π/180)

x0 = - model.l23 * cos(θ2) - model.l12 * cos(θ1)
y0 = - model.l23 * sin(θ2) - model.l12 * sin(θ1)

q0 = [x0, y0, θ1, θ2, θ3, θ4, θ5]
q̇0 = zeros(7)

A = 10
ω = 2π
# u = _t -> [0, A*sin(ω*_t), 0, 0]
u = _t -> [0,0,0,0]

joints = biped5link_kinematics(q0, model)

fpos = joints[1:2,:]

h = 0.01
tf = 10
T = Int.(tf/h)+1
times = h .* (0:T-1)
q_hist, q̇_hist, λ_hist = simulate(q0, q̇0, u, model, fpos, left_foot_constraint, h, T);

plot()
[plot!(times, q_hist[:,i], label="q$i") for i in 1:7]
display(plot!())
```
